    --> map({dict}) in map() method we simply pass a dict containing the key of value present in column and value of desired o/p.
            {'key':value,'key':value} - map({'female':0,'male':1}) - {'value present in column':desired value,
                                                                        'value present in column':desired value}

```python
df8 = pd.read_csv(r'Who_Suicide_Statistics.csv',nrows=15)  ## Reinitalizing the df8 since df8['sex'] is changed/modified...
```

#### Reinitalizing the df8 since df8['sex'] is changed/modified...

```python
df8.head()
```

```python
def modifySex(value):
    if(value == 'female'):
        return 0
    else:
        return 1
df8['sex'] = df8['sex'].apply(modifySex)
df8[:]
```

    --->>> apply() takes a function as its parameter. So i have declared a function(modifySex) and passed it to apply 
            function.

### Answer 4

        --> For replacing -1 with NaN's we have to use a parameter na_values while reading data...

```python
df8 = pd.read_csv(r'Who_Suicide_Statistics.csv',na_values=["-1"])
df8[:]
```

#### All the -ve values now changed to NaN values.
        --->>> The data obtained may not be clean always. Some of the data may have missing values, or some may have 
                Irrelevent data(Here No. of Suicides will not be -1 in any case). This Irrelevent Information is need to be 
                handled Here I want to Read the -1 data has NaN's.... This can be done by using 'na_values' parameter. 
                For example, If we pass 'male' as one of the na_value then all the items with 'male' is changed to NaN's.
                For Demonstration Lets Pass male is also one of the na_value....

```python
df8 = pd.read_csv(r'Who_Suicide_Statistics.csv',na_values=["-1","male"])
df8[:]
```

        -- > All the cells containing 'male' as the value are now turned into NaN's.

### Answer 5

        --> Our Aim here is to fill all the NaN Values present in suicide_no to 0 and the NaN Values present in population
            to the mean value of the population column.
            
        --> To fill all the NaN values Pandas has a method called fillna(). It can directly applied to entire dataframe 
            object or to a single Series/Column.
                Method #1) dataframe_object.fillna(values);
                Method #2) dataframe_object['Column_Name'].fillna(value);
                
                --> In first Method we can fill NaN values of every column at one go.
                --> In second Method we can fill NaN values of one specified column.

```python
df8.fillna({'sex' : "male",'suicides_no' : 0,},inplace =True)
df8['population'] = df8['population'].fillna(df8.population.mean())
df8[:]
```
